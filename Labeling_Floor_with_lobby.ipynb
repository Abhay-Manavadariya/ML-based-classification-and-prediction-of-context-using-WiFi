```python
import pandas as pd
import numpy as np
import re
```

```python
df = pd.read_excel('/content/LAB_DATA_G2.xlsx')
df.head()
```

```python
df=df.iloc[:,2:]
```

```python
df.head()
```

```python
def assign_label(location):
    match = re.match(r"LAB_(\d)_.*?([A-Z]+)_(\d+)", location)
    if match:
        floor, section, number = match.groups()
        floor = int(floor)
        number = int(number)

        # Lobbies A_1 to A_12, B_1 to B_12
        if section in {"A", "B"} and 1 <= number <= 12:
            return floor

        # Rooms C_1 and C_2 to O_1 and O_2
        if section in {"C", "D", "E", "F", "G", "H", "I", "J", "K", "L", "M", "N", "O"} and 1 <= number <= 2:
            return floor + 3

        # Rooms P_1 to P_12 and Q_1 to Q_12
        if section in {"P", "Q"} and 1 <= number <= 12:
            return floor + 6

        # Rooms C_13, C_14 to O_13, O_14
        if section in {"C", "D", "E", "F", "G", "H", "I", "J", "K", "L", "M", "N", "O"} and 13 <= number <= 14:
            return floor + 9

    return None

# Apply the function to the Location column to create a new 'Label' column
df['Label'] = df['Location'].apply(assign_label).astype('Int64')
```
